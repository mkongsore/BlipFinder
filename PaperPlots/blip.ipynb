```python
import numpy as np
import pandas as pd
import scipy.constants as const
import matplotlib.pyplot as plt
import scipy.interpolate as interp
import matplotlib as mpl
```

```python
plt.rcdefaults()
fontsize = 14
from matplotlib import font_manager
from matplotlib import rcParams
from matplotlib import rc

cmfont = font_manager.FontProperties(fname=mpl.get_data_path() + '/fonts/ttf/cmr10.ttf')
plt.rcParams['font.serif']=cmfont.get_name()
params = {"ytick.color" : "black",
          "xtick.color" : "black",
          "axes.labelcolor" : "black",
          "axes.edgecolor" : "black",
          "font.size" : fontsize,
          "text.usetex" : False,
          "mathtext.fontset" : "cm",
          "font.family" : "serif",
          "font.serif" : [cmfont.get_name()]}
plt.rcParams['axes.unicode_minus']=False
plt.rcParams.update(params)
```

```python
def snr(blip):
    return 1/np.sqrt(blip)*(1-np.sqrt(1/(1+blip**2)))
```

```python
b_ar = np.arange(1, 1000, .1)
#plt.figure(figsize = (8,6))

plt.plot(b_ar, 1/np.sqrt(b_ar), color = 'tab:red', label = 'blip$^{-1/2}$')
plt.plot(b_ar, snr(b_ar), lw = 3, color = 'tab:blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('blippiness')
plt.ylabel('SNR [$\delta_{Max}\sqrt{N_{obs}}$]')
plt.legend()
plt.savefig('flat.pdf', bbox_inches = 'tight')
```
